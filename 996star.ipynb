```python
from pyecharts import Line
import pandas as pd


data = pd.read_csv("./996stars.txt", header=None)[::50]
data[1] = data[1] * 1000 + data[2]
time_list = data[0].tolist()
data_list = data[1].tolist()

line = Line("996 star", background_color="#fff", width=1300)
line.add(None,time_list,data_list,is_smooth=True, is_datazoom_show=True, datazoom_range=[0, 100])
```
